# Relatório 2

## Lista de Tarefas

1. Mostrar quais bairros possui imóveis para locação e a quantidade de imóveis para locação.
2. Mostrar os 5 aluguéis mais caros dos bairros: Barra da Tijuca, Copacabana, Centro, Ipanema.
3. Criar uma nova coluna chamada Valor Total, para a soma do Aluguel, Condominio e IPTU.
4. Mostrar o valor total de Condomínio e IPTU para todos os bairros.

## Análise do Data-Base

```python
#Importando as bibliotecas
import pandas as pd
```

```python
#Importando o banco de dados.
dados = pd.read_csv('dbs/alugueis_sem_dados_nulos')
```

```python
#Visualização do banco
dados.head()
```

## 1 - Quais bairros possui imóveis para locação

```python
#Utilizando a função value_counts() é possível mostrar os bairros em ordem da quantidade de imóveis.
dados['Bairro'].value_counts()

```

## 2 - Aluguéis mais caros

```python
#Função que retorna os alugueis em ordem decrescente de Valor com parametro bairro
def alugueis_ordenados(bairro):
    alugueis_bairros = dados.query('Bairro == "{}"'.format(bairro))
    alugueis_ordenados = alugueis_bairros.sort_values(by=['Valor'], ascending=False)
    return alugueis_ordenados
```

### Barra da Tijuca

```python
#5 alugueis mais caros - Barra da Tijuca
alugueis_barra_da_tijuca = alugueis_ordenados("Barra da Tijuca")
alugueis_barra_da_tijuca.head(5)
```

### Copacabana

```python
#5 alugueis mais caros - Copacabana
alugueis_copacabana = alugueis_ordenados("Copacabana")
alugueis_copacabana.head(5)
```

### Centro

```python
#5 alugueis mais caros - Centro
alugueis_centro = alugueis_ordenados("Centro")
alugueis_centro.head(5)
```

### Ipanema

```python
#5 alugueis mais caros - Ipanema
alugueis_ipanema = alugueis_ordenados("Ipanema")
alugueis_ipanema.head(5)
```

## 3 - Criação da nova coluna Valor Total

```python
#Aqui é usado o metodo mais basico de insert de colunas do pandas
dados['Valor Total'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head()
```

## 4 - Soma dos IPTU's e Condomínios para todos os bairros

```python
#Função para soma do valor total de IPTU e Condominio para cada bairro
def soma_condominio_iptu(condominio_iptu):
    # Criando um array de bairros
    bairro = dados['Bairro'].unique() 
    soma = []
    # Laço para percorrer todo o array de bairro
    for i in range(len(bairro)):
        aux = dados.query(f'Bairro == "{bairro[i]}"')
        # Adicionando a um vetor a soma de condominio ou iptu para cada bairro do array
        soma.append(aux[f'{condominio_iptu}'].sum())

    # Criando um data frame das listas de bairro e de valor da soma
    aux = pd.DataFrame(list(zip(bairro, soma)), columns=['Bairro', f'Valor Total do {condominio_iptu}'])
    return aux
    
```

### IPTU

```python
#Chamando a função de soma para o IPTU
soma_iptu = soma_condominio_iptu('IPTU')
soma_iptu.head()
```

### Condomínio

```python
#Chamando a função de soma para o Condominio
soma_condominio = soma_condominio_iptu('Condominio')
soma_condominio.head()
```

```python
#Criação dos arquivos .csv dos dataframes obtidos
dados.to_csv(r'dbs/aluguel_coluna_valor_total', index = False)
soma_iptu.to_csv(r'dbs/bairro_soma_iptu', index = False)
soma_condominio.to_csv(r'dbs/bairro_soma_condominio', index = False)
```

Com essas duas lista montadas é feito um DataFrame com as colunas: Bairro e Valor Total.
